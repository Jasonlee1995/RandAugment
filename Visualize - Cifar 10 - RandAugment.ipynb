```python
import RandAugment, torch, torchvision
import matplotlib.pyplot as plt

# Ignore Warning
import warnings
warnings.filterwarnings(action='ignore')
```

```python
dataset = torchvision.datasets.CIFAR10('../../../../data/OCR/Jason/cifar10/', train=False, download=False)
```

```python
ex1, ex2, ex3, ex4 = dataset[2], dataset[25], dataset[28], dataset[42]
```

```python
aug_list = [[RandAugment.RandAug(1, 4, 1, 1), RandAugment.RandAug(3, 4, 1, 1)], 
            [RandAugment.RandAug(1, 5, 1, 1), RandAugment.RandAug(3, 5, 1, 1)], 
            [RandAugment.RandAug(1, 7, 1, 1), RandAugment.RandAug(3, 7, 1, 1)], 
            [RandAugment.RandAug(1, 9, 1, 1), RandAugment.RandAug(3, 9, 1, 1)], 
            [RandAugment.RandAug(1, 11, 1, 1), RandAugment.RandAug(3, 11, 1, 1)]]

title_list = [['N : 1, M : 4',  'N : 3, M : 4'],
              ['N : 1, M : 5',  'N : 3, M : 5'],
              ['N : 1, M : 7',  'N : 3, M : 7'],
              ['N : 1, M : 9',  'N : 3, M : 9'],
              ['N : 1, M : 11', 'N : 3, M : 11']]
```

```python
for i in range(len(aug_list)):
    for j in range(len(aug_list[i])):
        print(aug_list[i][j])
```

```python
plt.figure(figsize=(15, 40))
plt.rcParams['axes.titlesize'] = 20
plt.axis('off')

for i in range(len(aug_list)):
    plt.subplot(8, 3, 3*i+1)
    image, label = ex1
    img = plt.imshow(image); img.axes.get_xaxis().set_visible(False); img.axes.get_yaxis().set_visible(False)
    plt.gca().set_title('GT')
    
    for j in range(len(aug_list[i])):
        plt.subplot(8, 3, 3*i + j + 2)
        aug_image = aug_list[i][j](image)
        aug_image = plt.imshow(aug_image)
        aug_image.axes.get_xaxis().set_visible(False); aug_image.axes.get_yaxis().set_visible(False)
        plt.gca().set_title(title_list[i][j])
```
